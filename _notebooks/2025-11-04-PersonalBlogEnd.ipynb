## END OF TRIMESTER  BLOG

At the start of the year, I didn't know how to set up a code, open a terminal, or work in terminal. But now I know how to use Ubuntu, set up WSL, and make games like the one I made a couple weeks ago. At the start of the year, I had trouble understanding how to work in terminal and didn't like it because it didn't have buttons. But now, after all this class, I like working in terminology. It's so easy, simple, and more effective than working in GUI. VS Code used to be my enemy. I used to hate working in VS Code because it's so complicated, so many files. It's just not what I feel like is an ideal working space. But now, it's different because VS Code really allows you to manage what you're working on, use Terminator files, and run some pretty key jokes. 

Some of the main learning things that I learned were setting up WSL and Ubuntu on the terminal. After that, I learned some basics of JavaScript, like:
- Variables
- Classes
- Arrays
- Some mathematical expressions in JavaScript


And then we learned how to apply it. We made a game, each group having their own level, and we made a staircase game where you walk the staircase and dodge skeletons. We used countless classes, variables, and arrays for the levels. 

Night in the museum was an amazing time. We showed people how our game worked, gave them a quick demo, and let them play. They also gave some positive feedback and some things to work on. Here are two positive feedbacks and one things that we need to improve.

If I were to restart, I would stop relying on AI as much as I did in this project. Split the tasks up and give AI smaller tasks while understanding what it's doing is critical for making our game better. I want to make this again. I would definitely go step-by-step and make this non-buggy and as efficient as possible. Make it smooth and flowing. And since we would have more time, I would have time to think and add more features in. Again, in those baby steps. 

I'm not really sure what I want to learn in the next co-op sci. I just want to know how to work on computers better, maybe start some web development because that's really cool. In the modern world, knowing how to make websites and some back-end related to that would be a pretty cool tool. Just saying.
